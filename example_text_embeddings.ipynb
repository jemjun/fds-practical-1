```python
from sentence_transformers import SentenceTransformer
import numpy as np
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

# Text Embeddings

Text embeddings are numerical representations of text that capture semantic meaning in a high-dimensional vector space, where similar concepts are positioned closer together. They're created by machine learning models that map words, sentences, or documents to dense vectors of real numbersâ€”typically hundreds or thousands of dimensions. These embeddings enable computers to understand relationships between pieces of text and power applications like semantic search, recommendation systems, and clustering similar content.

```python
desc_1 = "The company produces cars."
desc_2 = "The company manufactures automotives."
desc_3 = "The company operates grocery stores."
```

```python
emb_1 = model.encode(desc_1)
emb_2 = model.encode(desc_2)
emb_3 = model.encode(desc_3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define two vectors
v1 = np.array([3, 4])
v2 = np.array([3, 4])
v2 = np.array([4, 1])
#v2 = np.array([-3, -4])

# Calculate cosine similarity
cos_sim = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
angle = np.arccos(cos_sim) * 180 / np.pi

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Left plot: Vector visualization
ax1.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.006, label='v1')
ax1.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.006, label='v2')
ax1.set_xlim(-1, 6)
ax1.set_ylim(-1, 5)
ax1.set_aspect('equal')
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='k', linewidth=0.5)
ax1.axvline(x=0, color='k', linewidth=0.5)
ax1.set_title(f'Vectors with angle = {angle:.1f}Â°')
ax1.legend()

# Right plot: Cosine similarity as bar
ax2.bar(['Cosine Similarity'], [cos_sim], color='green', alpha=0.7)
ax2.set_ylim(-1, 1)
ax2.axhline(y=0, color='k', linewidth=0.5)
ax2.set_ylabel('Similarity')
ax2.set_title(f'Cosine Similarity = {cos_sim:.3f}')
ax2.grid(True, alpha=0.3)

plt.suptitle('Cosine Similarity Visualization', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Print values
print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Cosine Similarity: {cos_sim:.3f}")
print(f"Angle between vectors: {angle:.1f}Â°")
```

```python
def cosine_similarity(vec_1, vec_2):
    cos_sim = np.dot(vec_1, vec_2) / (np.linalg.norm(vec_1) * np.linalg.norm(vec_2))
    return cos_sim
```

```python
print(cosine_similarity(emb_1, emb_2))
print(cosine_similarity(emb_1, emb_3))
```
